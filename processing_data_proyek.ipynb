```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Mengatur gaya visualisasi
sns.set(style="whitegrid")
```

```python
# Membaca file CSV
data = pd.read_csv('Assignment-1_Data.csv', sep=';')
```

```python
# Melihat beberapa baris pertama
print(data.head())

# Melihat informasi umum tentang dataset
print(data.info())

# Melihat statistik deskriptif
print(data.describe())
```

```python
# Menghitung missing values
missing_values = data.isnull().sum()
print(missing_values)

# Visualisasi missing values
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Values Heatmap')
plt.show()
```

```python
# Menghapus baris dengan missing values
data.dropna(inplace=True)
```

```python
#Ubah tipe data kolom yang diperlukan
data['Price'] = data['Price'].str.replace(',', '.').astype(float)
```

```python
#Visualisasikan distribusi dari kolom numerik, seperti harga
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=30, kde=True)
plt.title('Distribusi Harga')
plt.xlabel('Harga')
plt.ylabel('Frekuensi')
plt.show()
```

```python
#Analisis Kategorikal
#Visualisasikan jumlah item berdasarkan negara.
plt.figure(figsize=(12, 6))
sns.countplot(y='Country', data=data, order=data['Country'].value_counts().index)
plt.title('Jumlah Item Berdasarkan Negara')
plt.xlabel('Jumlah Item')
plt.ylabel('Negara')
plt.show()
```

```python
#Korelasi Antara Variabel
#Analisis korelasi antar variabel numerik.
plt.figure(figsize=(10, 8))
# Select only numerical features for correlation calculation
numerical_data = data.select_dtypes(include=np.number)
correlation_matrix = numerical_data.corr()
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matriks Korelasi')
plt.show()
```
